## Introduction to K-Nearest Neighbors (KNN)

**K-Nearest Neighbors (KNN)** is a fundamental machine learning algorithm that embodies the principle of **"learning by analogy"**. The core idea is beautifully simple: *to classify a new data point, look at the K closest examples in the training data and let them vote on the classification*.

### ü§î The Intuitive Appeal

Imagine you're trying to decide whether to watch a new movie. What would you naturally do?
- **Look at similar movies** you've enjoyed before
- **Ask friends with similar tastes** for their opinions
- **Consider the genre, director, actors** that you usually like

This is exactly how KNN works! It finds the **K most similar** training examples and uses their labels to make a prediction.

### üîç Real-World Analogy

**Medical Diagnosis Example:**
- **Problem**: Diagnose a new patient's condition
- **Approach**: Find K patients with similar symptoms, age, medical history
- **Decision**: Use the most common diagnosis among these similar cases
- **K=5**: If 4 out of 5 similar patients had Disease A, predict Disease A

### üìä Key Characteristics

#### **1. Non-Parametric Algorithm**
- **No assumptions** about the underlying data distribution
- **No training phase** - simply stores all training data
- **Flexible decision boundaries** that adapt to data complexity

#### **2. Instance-Based Learning**
- **Lazy learning**: No explicit training phase
- **Memory-based**: Stores entire training dataset
- **Local learning**: Decisions based on local neighborhoods

#### **3. Supervised Learning**
- **Uses labeled training data** for classification
- **Direct mapping** from features to class labels
- **Non-linear decision boundaries** naturally emerge

### üéØ Fundamental Concept

The effectiveness of KNN relies on the **assumption of locality**:
> *"Similar objects tend to belong to the same class"*

This means:
- **Objects close in feature space** likely share the same label
- **Distance metrics** determine what "close" means
- **Local patterns** in data drive classification decisions

### üîÑ Algorithm Philosophy

Unlike other algorithms that learn a global model:

#### **Traditional ML Approach:**
1. **Training Phase**: Learn parameters from training data
2. **Prediction Phase**: Apply learned model to new data

#### **KNN Approach:**
1. **"Training" Phase**: Simply store all training examples
2. **Prediction Phase**: 
   - Find K nearest neighbors to new point
   - Use their labels to classify the new point

### üåü Why KNN is Special

#### **Simplicity**
- **Easy to understand** and explain to stakeholders
- **No complex mathematics** or optimization required
- **Intuitive decision-making process**

#### **Flexibility**
- **Adapts to any data distribution** (linear, non-linear, complex)
- **No assumptions** about feature relationships
- **Works with different data types** (numerical, categorical with proper encoding)

#### **Effectiveness**
- **Often competitive** with more complex algorithms
- **Good baseline** for comparison with other methods
- **Robust to outliers** when K > 1

### üìà When KNN Shines

KNN is particularly effective when:
- **Local patterns** are more important than global trends
- **Decision boundaries** are irregular or complex
- **Training data** is abundant and representative
- **Interpretability** is important for decision-making
- **Quick prototyping** is needed

### üîç The Big Picture

KNN represents a different philosophy in machine learning:
- **Instance-based** rather than model-based learning
- **Local** rather than global decision making
- **Similarity-driven** rather than parameter-driven predictions

This makes KNN an excellent starting point for understanding classification problems and a powerful tool in the machine learning toolkit.

## How KNN Works: Theory and Mathematical Foundation

### üî¢ The KNN Algorithm Step-by-Step

The KNN algorithm follows a straightforward process for classifying new data points:

#### **Step 1: Store Training Data**
- Keep all training examples: **{(x‚ÇÅ, y‚ÇÅ), (x‚ÇÇ, y‚ÇÇ), ..., (x‚Çô, y‚Çô)}**
- No model parameters to learn
- Simply maintain the dataset in memory

#### **Step 2: Calculate Distances**
For a new point **x_new**:
- Calculate distance from **x_new** to every training point **x·µ¢**
- Use a chosen distance metric (Euclidean, Manhattan, etc.)

#### **Step 3: Find K Nearest Neighbors**
- Sort all training points by distance to **x_new**
- Select the **K closest** points
- These form the "neighborhood" for prediction

#### **Step 4: Make Prediction**
- **Classification**: Take majority vote from K neighbors
- **Regression**: Take average of K neighbors' values

### üìè Distance Metrics: The Heart of KNN

The choice of distance metric fundamentally affects KNN performance:

#### **1. Euclidean Distance** (Most Common)
```
d(x, y) = ‚àö[(x‚ÇÅ-y‚ÇÅ)¬≤ + (x‚ÇÇ-y‚ÇÇ)¬≤ + ... + (x‚Çô-y‚Çô)¬≤]
```
- **Best for**: Continuous features, when all features are equally important
- **Geometric interpretation**: Straight-line distance in n-dimensional space
- **Sensitivity**: Affected by feature scales - requires normalization

#### **2. Manhattan Distance** (City Block)
```
d(x, y) = |x‚ÇÅ-y‚ÇÅ| + |x‚ÇÇ-y‚ÇÇ| + ... + |x‚Çô-y‚Çô|
```
- **Best for**: High-dimensional data, when path length matters more than direct distance
- **Geometric interpretation**: Distance along grid lines (like city blocks)
- **Robustness**: Less sensitive to outliers than Euclidean

#### **3. Minkowski Distance** (Generalization)
```
d(x, y) = (|x‚ÇÅ-y‚ÇÅ|·µñ + |x‚ÇÇ-y‚ÇÇ|·µñ + ... + |x‚Çô-y‚Çô|·µñ)^(1/p)
```
- **p=1**: Manhattan distance
- **p=2**: Euclidean distance
- **p=‚àû**: Chebyshev distance (maximum difference across dimensions)

#### **4. Hamming Distance**
```
d(x, y) = number of positions where x·µ¢ ‚â† y·µ¢
```
- **Best for**: Categorical features, binary data
- **Use case**: Text classification, DNA sequence analysis

### üéØ The Voting Mechanism

#### **Simple Majority Vote**
```
Class = argmax(count of each class among K neighbors)
```
- Each neighbor gets equal vote weight
- Ties can be broken randomly or by decreasing K

#### **Distance-Weighted Voting**
```
Weight(i) = 1 / distance(i)
Class = argmax(Œ£ weights for each class)
```
- Closer neighbors have more influence
- Prevents ties and provides smoother decision boundaries
- More stable predictions

### üîë Key Parameters

#### **1. K (Number of Neighbors)**

**Small K (K=1, K=3):**
- **Advantages**: Sensitive to local patterns, flexible decision boundaries
- **Disadvantages**: Prone to noise, overfitting, unstable predictions
- **Use when**: Low noise data, need to capture fine details

**Large K (K=20, K=50):**
- **Advantages**: Stable predictions, less sensitive to noise, smoother boundaries
- **Disadvantages**: May miss local patterns, underfitting, computational cost
- **Use when**: Noisy data, need stable predictions

**Optimal K Selection:**
- **Odd numbers**: Avoid ties in binary classification
- **Cross-validation**: Test different K values systematically
- **Rule of thumb**: K = ‚àön (where n = training samples)
- **Domain knowledge**: Consider noise level and pattern complexity

#### **2. Distance Metric Choice**

**Factors to Consider:**
- **Data type**: Continuous vs. categorical features
- **Feature scales**: Whether features are normalized
- **Dimensionality**: High-dimensional data behavior
- **Domain knowledge**: Physical meaning of "similarity"

### üîÑ Decision Boundary Properties

KNN creates **non-linear, adaptive decision boundaries**:

#### **Characteristics:**
- **Piecewise linear** segments forming complex shapes
- **Data-driven**: Boundaries adapt to training data distribution
- **Local**: Different regions can have different boundary orientations
- **Irregular**: Can handle complex, multi-modal class distributions

#### **Boundary Smoothness:**
- **Low K**: Jagged, highly flexible boundaries (potential overfitting)
- **High K**: Smoother, more regular boundaries (potential underfitting)
- **Optimal K**: Balance between flexibility and smoothness

This mathematical foundation shows why KNN is both powerful and intuitive - it directly leverages the geometric structure of the data to make predictions!

## Pros and Cons of K-Nearest Neighbors

Understanding the strengths and limitations of KNN is crucial for knowing when and how to apply this algorithm effectively.

### ‚úÖ Advantages of KNN

#### **1. Simplicity and Intuitive Appeal**
- **Easy to understand**: The logic mirrors human decision-making
- **No complex mathematics**: No derivatives, gradients, or optimization
- **Quick to implement**: Few lines of code for basic implementation
- **Transparent process**: Easy to explain predictions to stakeholders

#### **2. No Training Period Required**
- **Instant availability**: Ready to make predictions immediately
- **No model fitting**: Saves time in development phase
- **Dynamic learning**: New data points immediately affect predictions
- **Online learning capability**: Can easily incorporate new examples

#### **3. Handles Non-Linear Relationships**
- **Complex decision boundaries**: Naturally adapts to data shape
- **No linearity assumptions**: Works with any data distribution
- **Local adaptability**: Different regions can follow different patterns
- **Flexible classification**: Can handle multi-modal class distributions

#### **4. No Data Distribution Assumptions**
- **Non-parametric**: Works regardless of underlying data distribution
- **Robust to outliers**: When K > 1, outliers have limited impact
- **Universal applicability**: Works with various types of problems
- **Distribution-free**: No normality or other statistical assumptions

#### **5. Naturally Handles Multi-Class Problems**
- **Direct multi-class support**: No need for one-vs-rest strategies
- **Scales to many classes**: Works equally well with 2 or 20 classes
- **Class probability estimates**: Can provide confidence measures
- **Balanced multi-class handling**: No inherent bias toward majority classes

#### **6. Effective for Small Datasets**
- **Low data requirements**: Can work with minimal training data
- **No overfitting with small samples**: Unlike complex models
- **Preserves all information**: Uses every training example
- **Good baseline performance**: Often competitive with simple datasets

---

### ‚ùå Disadvantages of KNN

#### **1. Computational Intensity**
- **Expensive prediction phase**: Must calculate distances to all training points
- **O(n) complexity per prediction**: Scales poorly with large datasets
- **Memory requirements**: Must store entire training dataset
- **Real-time challenges**: Slow predictions for large datasets

**Impact:**
- **Large datasets**: Prediction time becomes prohibitive
- **Real-time applications**: May not meet latency requirements
- **Resource constraints**: High memory and CPU usage

#### **2. Curse of Dimensionality**
- **Distance becomes meaningless**: In high dimensions, all points seem equidistant
- **Nearest neighbors become distant**: Concept of "nearness" breaks down
- **Performance degradation**: Accuracy drops as dimensions increase
- **Concentration phenomenon**: Distances concentrate around their mean

**Mathematical Insight:**
In high dimensions (d > 10-20):
- **All distances become similar**: Hard to distinguish near from far
- **Volume concentration**: Most volume is near the boundary of hyperspheres
- **Irrelevant features dominate**: Noise dimensions overwhelm signal

#### **3. Sensitivity to Feature Scaling**
- **Scale dependency**: Features with larger ranges dominate distance calculations
- **Requires preprocessing**: Must normalize or standardize features
- **Unit sensitivity**: Results change with measurement units
- **Preprocessing overhead**: Additional data preparation steps required

**Example Problem:**
- Feature 1: Age (0-100)
- Feature 2: Income ($20,000-$200,000)
- Income differences will overwhelm age differences in distance calculations

#### **4. Poor Performance with Irrelevant Features**
- **Noise amplification**: Irrelevant features add noise to distance calculations
- **Feature selection critical**: Requires careful feature engineering
- **Dimensional dilution**: Important features get diluted by noise
- **Manual intervention needed**: Algorithm doesn't automatically identify important features

#### **5. Sensitive to Local Structure**
- **Uneven class distributions**: Performance varies across different regions
- **Boundary sensitivity**: Struggles near class boundaries
- **Local sparsity**: Poor performance in sparse regions of feature space
- **Imbalanced data issues**: Majority classes can dominate neighborhoods

#### **6. Choice of K is Critical**
- **No universal optimal K**: Requires dataset-specific tuning
- **Bias-variance trade-off**: Must balance underfitting vs. overfitting
- **Cross-validation needed**: Systematic approach required for K selection
- **Domain-dependent**: Optimal K varies with problem characteristics

#### **7. Storage and Memory Requirements**
- **Lazy learning cost**: Must store entire training dataset
- **Memory scaling**: Storage requirements grow linearly with data size
- **No model compression**: Cannot reduce to a compact representation
- **Deployment challenges**: Large models difficult to deploy

---

### ‚öñÔ∏è Trade-offs Summary

| **Aspect** | **Advantage** | **Disadvantage** |
|------------|---------------|------------------|
| **Learning** | No training required | Expensive prediction phase |
| **Complexity** | Simple to understand | Computationally intensive |
| **Flexibility** | Handles any distribution | Sensitive to irrelevant features |
| **Scalability** | Good for small data | Poor for large datasets |
| **Dimensionality** | No assumptions needed | Curse of dimensionality |
| **Memory** | Preserves all information | High storage requirements |

---

### üéØ When to Use KNN

#### **‚úÖ KNN is Excellent When:**
- **Small to medium datasets** (< 10,000 samples)
- **Low to moderate dimensions** (< 20 features)
- **Complex, non-linear patterns** that are hard to model parametrically
- **Interpretability is important** and you need to explain decisions
- **Quick prototyping** or establishing baseline performance
- **Local patterns** are more important than global trends

#### **‚ùå Avoid KNN When:**
- **Large datasets** with millions of samples
- **High-dimensional data** without proper dimensionality reduction
- **Real-time predictions** with strict latency requirements
- **Many irrelevant features** without feature selection
- **Severe class imbalance** without proper handling
- **Memory constraints** are tight

---

### üí° Optimization Strategies

To mitigate KNN's disadvantages:

#### **1. Computational Optimization**
- **Approximate nearest neighbors**: Use algorithms like LSH, k-d trees
- **Dimensionality reduction**: PCA, t-SNE before applying KNN
- **Instance selection**: Remove redundant training examples

#### **2. Feature Engineering**
- **Feature scaling**: Always normalize or standardize features
- **Feature selection**: Remove irrelevant and redundant features
- **Domain-specific distances**: Use appropriate distance metrics

#### **3. Algorithm Variants**
- **Weighted KNN**: Distance-weighted voting for better predictions
- **Adaptive K**: Use different K values for different regions
- **Local outlier detection**: Remove noisy training examples

Understanding these trade-offs helps you make informed decisions about when and how to apply KNN effectively!

````python
# Exploratory Data Analysis for KNN Considerations
print("üîç EXPLORATORY DATA ANALYSIS FOR KNN")
print("=" * 60)

# Function to analyze dataset characteristics important for KNN
def analyze_dataset_for_knn(X, y, dataset_name, feature_names=None):
    """Analyze dataset characteristics relevant for KNN performance"""
    print(f"\nüìä {dataset_name.upper()} DATASET ANALYSIS:")
    print("-" * 50)
    
    # Basic statistics
    print(f"Shape: {X.shape} (samples √ó features)")
    print(f"Classes: {len(np.unique(y))} unique classes")
    print(f"Class distribution: {dict(zip(*np.unique(y, return_counts=True)))}")
    
    # Feature scaling analysis
    feature_ranges = []
    feature_means = []
    feature_stds = []
    
    for i in range(X.shape[1]):
        feature_ranges.append(X[:, i].max() - X[:, i].min())
        feature_means.append(X[:, i].mean())
        feature_stds.append(X[:, i].std())
    
    print(f"\nFeature Scale Analysis:")
    print(f"   ‚Ä¢ Range spans: {min(feature_ranges):.2f} to {max(feature_ranges):.2f}")
    print(f"   ‚Ä¢ Mean values: {min(feature_means):.2f} to {max(feature_means):.2f}")
    print(f"   ‚Ä¢ Std deviations: {min(feature_stds):.2f} to {max(feature_stds):.2f}")
    
    # Check if scaling is needed
    range_ratio = max(feature_ranges) / min(feature_ranges) if min(feature_ranges) > 0 else float('inf')
    if range_ratio > 10:
        print(f"   ‚ö†Ô∏è SCALING NEEDED: Range ratio = {range_ratio:.1f} (>10)")
    else:
        print(f"   ‚úÖ Scaling beneficial: Range ratio = {range_ratio:.1f}")
    
    # Dimensionality assessment
    n_samples, n_features = X.shape
    if n_features > 20:
        print(f"   ‚ö†Ô∏è HIGH DIMENSIONAL: {n_features} features may suffer from curse of dimensionality")
    elif n_features > 10:
        print(f"   üî∏ MODERATE DIMENSIONS: {n_features} features - monitor performance")
    else:
        print(f"   ‚úÖ LOW DIMENSIONS: {n_features} features - good for KNN")
    
    # Sample size assessment
    samples_per_feature = n_samples / n_features
    if samples_per_feature < 10:
        print(f"   ‚ö†Ô∏è LIMITED DATA: {samples_per_feature:.1f} samples per feature")
    else:
        print(f"   ‚úÖ ADEQUATE DATA: {samples_per_feature:.1f} samples per feature")
    
    return {
        'range_ratio': range_ratio,
        'n_features': n_features,
        'samples_per_feature': samples_per_feature,
        'feature_ranges': feature_ranges
    }

# Analyze all datasets
iris_analysis = analyze_dataset_for_knn(iris.data, iris.target, "Iris", iris.feature_names)
wine_analysis = analyze_dataset_for_knn(wine.data, wine.target, "Wine", wine.feature_names)
digits_analysis = analyze_dataset_for_knn(digits_subset_X, digits_subset_y, "Digits")

# Detailed feature scaling visualization for Iris (most interpretable)
print(f"\nüìà FEATURE SCALING VISUALIZATION (Iris Dataset):")
print("-" * 60)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Before scaling - feature distributions
for i, feature in enumerate(iris.feature_names):
    ax = axes[i//2, i%2]
    for class_idx, class_name in enumerate(iris.target_names):
        mask = iris.target == class_idx
        ax.hist(iris.data[mask, i], alpha=0.6, label=class_name, bins=15)
    
    ax.set_title(f'{feature}\n(Range: {iris.data[:, i].min():.1f} - {iris.data[:, i].max():.1f})', 
                 fontweight='bold')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.suptitle('Iris Dataset: Feature Distributions by Class\n(Before Scaling)', 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Show feature scaling necessity
print(f"\n‚öñÔ∏è SCALING NECESSITY SUMMARY:")
print("-" * 40)
print(f"Iris Dataset:")
print(f"   ‚Ä¢ Range ratio: {iris_analysis['range_ratio']:.1f}")
print(f"   ‚Ä¢ Recommendation: {'Scaling required' if iris_analysis['range_ratio'] > 10 else 'Scaling beneficial'}")

print(f"\nWine Dataset:")
print(f"   ‚Ä¢ Range ratio: {wine_analysis['range_ratio']:.1f}")
print(f"   ‚Ä¢ Recommendation: {'Scaling required' if wine_analysis['range_ratio'] > 10 else 'Scaling beneficial'}")

print(f"\nDigits Dataset:")
print(f"   ‚Ä¢ Range ratio: {digits_analysis['range_ratio']:.1f}")
print(f"   ‚Ä¢ Recommendation: {'Scaling required' if digits_analysis['range_ratio'] > 10 else 'Scaling beneficial'}")

# Sample complexity analysis
print(f"\nüéØ KNN SUITABILITY ASSESSMENT:")
print("-" * 50)

datasets = ['Iris', 'Wine', 'Digits']
analyses = [iris_analysis, wine_analysis, digits_analysis]

for dataset, analysis in zip(datasets, analyses):
    print(f"\n{dataset} Dataset:")
    
    # Dimensionality score
    if analysis['n_features'] <= 10:
        dim_score = "‚úÖ Excellent"
    elif analysis['n_features'] <= 20:
        dim_score = "üî∏ Good"
    else:
        dim_score = "‚ö†Ô∏è Challenging"
    
    # Data adequacy score  
    if analysis['samples_per_feature'] >= 20:
        data_score = "‚úÖ Excellent"
    elif analysis['samples_per_feature'] >= 10:
        data_score = "üî∏ Adequate"
    else:
        data_score = "‚ö†Ô∏è Limited"
    
    # Scaling requirement
    if analysis['range_ratio'] <= 5:
        scale_score = "‚úÖ Minimal scaling needed"
    elif analysis['range_ratio'] <= 50:
        scale_score = "üî∏ Scaling recommended"
    else:
        scale_score = "‚ö†Ô∏è Scaling critical"
    
    print(f"   ‚Ä¢ Dimensionality: {dim_score} ({analysis['n_features']} features)")
    print(f"   ‚Ä¢ Data adequacy: {data_score} ({analysis['samples_per_feature']:.1f} samples/feature)")
    print(f"   ‚Ä¢ Scaling needs: {scale_score} (ratio: {analysis['range_ratio']:.1f})")

print(f"\nüî¨ KEY INSIGHTS FOR KNN IMPLEMENTATION:")
print(f"   1. All datasets will benefit from feature scaling")
print(f"   2. Iris: Ideal for KNN (low dim, balanced, adequate data)")
print(f"   3. Wine: Good for KNN but needs careful scaling")
print(f"   4. Digits: Challenging due to high dimensionality")
print(f"   5. Feature scaling is critical for all datasets")
print(f"   6. Consider dimensionality reduction for Digits dataset")
````

## Model Building: KNN in Python with Scikit-Learn

Now we'll implement KNN classifiers for each dataset, demonstrating proper preprocessing, parameter tuning, and model evaluation.

````python
# Comprehensive KNN Model Building Pipeline
print("üèóÔ∏è KNN MODEL BUILDING PIPELINE")
print("=" * 60)

# Function to build and evaluate KNN models
def build_knn_model(X, y, dataset_name, feature_names=None, test_size=0.2, random_state=42):
    """
    Complete KNN model building pipeline with preprocessing and evaluation
    """
    print(f"\nüöÄ BUILDING KNN MODEL FOR {dataset_name.upper()}")
    print("-" * 50)
    
    # Step 1: Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )
    
    print(f"üìä Data Split:")
    print(f"   ‚Ä¢ Training: {X_train.shape[0]} samples")
    print(f"   ‚Ä¢ Testing: {X_test.shape[0]} samples")
    print(f"   ‚Ä¢ Features: {X_train.shape[1]}")
    
    # Step 2: Feature Scaling (Critical for KNN)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print(f"\nüîß Feature Scaling Applied:")
    print(f"   ‚Ä¢ Method: StandardScaler (mean=0, std=1)")
    print(f"   ‚Ä¢ Before scaling - feature ranges: {X_train.min(axis=0).min():.2f} to {X_train.max(axis=0).max():.2f}")
    print(f"   ‚Ä¢ After scaling - feature ranges: {X_train_scaled.min():.2f} to {X_train_scaled.max():.2f}")
    
    # Step 3: Basic KNN Model (K=5 as starting point)
    knn_basic = KNeighborsClassifier(n_neighbors=5)
    knn_basic.fit(X_train_scaled, y_train)
    
    # Basic predictions
    y_pred_basic = knn_basic.predict(X_test_scaled)
    basic_accuracy = accuracy_score(y_test, y_pred_basic)
    
    print(f"\nüéØ Basic KNN Model (K=5):")
    print(f"   ‚Ä¢ Test Accuracy: {basic_accuracy:.4f}")
    
    # Step 4: K Parameter Optimization using Cross-Validation
    print(f"\nüîç OPTIMIZING K PARAMETER:")
    print("-" * 30)
    
    # Test different K values
    k_values = list(range(1, 21))  # Test K from 1 to 20
    cv_scores = []
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        scores = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
        cv_scores.append(scores.mean())
    
    # Find optimal K
    optimal_k = k_values[np.argmax(cv_scores)]
    optimal_score = max(cv_scores)
    
    print(f"   ‚Ä¢ Tested K values: {min(k_values)} to {max(k_values)}")
    print(f"   ‚Ä¢ Optimal K: {optimal_k}")
    print(f"   ‚Ä¢ Cross-validation score: {optimal_score:.4f}")
    
    # Step 5: Distance Metric Optimization
    print(f"\nüìè OPTIMIZING DISTANCE METRIC:")
    print("-" * 35)
    
    distance_metrics = ['euclidean', 'manhattan', 'minkowski']
    distance_scores = {}
    
    for metric in distance_metrics:
        if metric == 'minkowski':
            # Test different p values for Minkowski
            for p in [1, 2, 3]:
                knn = KNeighborsClassifier(n_neighbors=optimal_k, metric=metric, p=p)
                scores = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
                distance_scores[f'{metric}_p{p}'] = scores.mean()
        else:
            knn = KNeighborsClassifier(n_neighbors=optimal_k, metric=metric)
            scores = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
            distance_scores[metric] = scores.mean()
    
    # Find best distance metric
    best_metric = max(distance_scores, key=distance_scores.get)
    best_metric_score = distance_scores[best_metric]
    
    print(f"   ‚Ä¢ Distance metrics tested: {list(distance_scores.keys())}")
    print(f"   ‚Ä¢ Best metric: {best_metric}")
    print(f"   ‚Ä¢ Best metric score: {best_metric_score:.4f}")
    
    # Display all distance metric results
    for metric, score in distance_scores.items():
        print(f"     - {metric}: {score:.4f}")
    
    # Step 6: Build Final Optimized Model
    print(f"\nüèÜ FINAL OPTIMIZED MODEL:")
    print("-" * 30)
    
    # Parse best metric for final model
    if 'minkowski' in best_metric:
        p_value = int(best_metric.split('_p')[1])
        final_knn = KNeighborsClassifier(
            n_neighbors=optimal_k, 
            metric='minkowski', 
            p=p_value
        )
    else:
        final_knn = KNeighborsClassifier(
            n_neighbors=optimal_k, 
            metric=best_metric
        )
    
    # Train final model
    final_knn.fit(X_train_scaled, y_train)
    
    # Final predictions
    y_pred_final = final_knn.predict(X_test_scaled)
    y_prob_final = final_knn.predict_proba(X_test_scaled)
    
    final_accuracy = accuracy_score(y_test, y_pred_final)
    
    print(f"   ‚Ä¢ Final K: {optimal_k}")
    print(f"   ‚Ä¢ Final Distance Metric: {best_metric}")
    print(f"   ‚Ä¢ Final Test Accuracy: {final_accuracy:.4f}")
    print(f"   ‚Ä¢ Improvement over basic: {final_accuracy - basic_accuracy:+.4f}")
    
    # Return all important results
    return {
        'dataset_name': dataset_name,
        'X_train_scaled': X_train_scaled,
        'X_test_scaled': X_test_scaled,
        'y_train': y_train,
        'y_test': y_test,
        'final_model': final_knn,
        'y_pred': y_pred_final,
        'y_prob': y_prob_final,
        'optimal_k': optimal_k,
        'best_metric': best_metric,
        'final_accuracy': final_accuracy,
        'k_values': k_values,
        'cv_scores': cv_scores,
        'distance_scores': distance_scores,
        'scaler': scaler
    }

# Build models for all datasets
print("üîÑ Building KNN models for all datasets...")

# Build Iris model
iris_results = build_knn_model(iris.data, iris.target, "Iris", iris.feature_names)

# Build Wine model  
wine_results = build_knn_model(wine.data, wine.target, "Wine", wine.feature_names)

# Build Digits model
digits_results = build_knn_model(digits_subset_X, digits_subset_y, "Digits")

print(f"\n‚úÖ ALL MODELS BUILT SUCCESSFULLY!")
print(f"üìä QUICK COMPARISON:")
print("-" * 30)
print(f"Iris    - Accuracy: {iris_results['final_accuracy']:.4f} (K={iris_results['optimal_k']}, {iris_results['best_metric']})")
print(f"Wine    - Accuracy: {wine_results['final_accuracy']:.4f} (K={wine_results['optimal_k']}, {wine_results['best_metric']})")
print(f"Digits  - Accuracy: {digits_results['final_accuracy']:.4f} (K={digits_results['optimal_k']}, {digits_results['best_metric']})")
````

````python
# Comprehensive Model Evaluation
print("üìä COMPREHENSIVE KNN MODEL EVALUATION")
print("=" * 60)

def evaluate_knn_model(results, target_names=None):
    """
    Comprehensive evaluation of KNN model performance
    """
    dataset_name = results['dataset_name']
    y_test = results['y_test']
    y_pred = results['y_pred']
    y_prob = results['y_prob']
    
    print(f"\nüéØ {dataset_name.upper()} DATASET EVALUATION:")
    print("-" * 50)
    
    # Basic Classification Metrics
    accuracy = accuracy_score(y_test, y_pred)
    
    # Handle multi-class metrics
    n_classes = len(np.unique(y_test))
    if n_classes == 2:
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        # For binary classification, use second class (index 1) for ROC-AUC
        roc_auc = roc_auc_score(y_test, y_prob[:, 1])
    else:
        # Multi-class metrics with macro averaging
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')
        # For multi-class ROC-AUC, use one-vs-rest approach
        roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr', average='macro')
    
    print(f"üìà CLASSIFICATION METRICS:")
    print(f"   ‚Ä¢ Accuracy:  {accuracy:.4f}")
    print(f"   ‚Ä¢ Precision: {precision:.4f} (macro avg)")
    print(f"   ‚Ä¢ Recall:    {recall:.4f} (macro avg)")
    print(f"   ‚Ä¢ F1-Score:  {f1:.4f} (macro avg)")
    print(f"   ‚Ä¢ ROC-AUC:   {roc_auc:.4f} (macro avg)")
    
    # Confusion Matrix Analysis
    cm = confusion_matrix(y_test, y_pred)
    print(f"\nüßÆ CONFUSION MATRIX:")
    print(cm)
    
    # Per-class analysis
    if target_names is not None:
        print(f"\nüìã DETAILED CLASSIFICATION REPORT:")
        report = classification_report(y_test, y_pred, target_names=target_names)
        print(report)
    else:
        print(f"\nüìã DETAILED CLASSIFICATION REPORT:")
        report = classification_report(y_test, y_pred)
        print(report)
    
    # Model confidence analysis
    prediction_confidence = np.max(y_prob, axis=1)
    print(f"\nüéØ PREDICTION CONFIDENCE ANALYSIS:")
    print(f"   ‚Ä¢ Mean confidence: {prediction_confidence.mean():.4f}")
    print(f"   ‚Ä¢ Min confidence:  {prediction_confidence.min():.4f}")
    print(f"   ‚Ä¢ Max confidence:  {prediction_confidence.max():.4f}")
    print(f"   ‚Ä¢ Std confidence:  {prediction_confidence.std():.4f}")
    
    # Identify low-confidence predictions
    low_confidence_threshold = 0.5
    low_confidence_mask = prediction_confidence < low_confidence_threshold
    n_low_confidence = low_confidence_mask.sum()
    
    if n_low_confidence > 0:
        print(f"   ‚ö†Ô∏è {n_low_confidence} predictions with confidence < {low_confidence_threshold}")
        low_conf_accuracy = accuracy_score(y_test[low_confidence_mask], y_pred[low_confidence_mask])
        print(f"   ‚Ä¢ Accuracy of low-confidence predictions: {low_conf_accuracy:.4f}")
    else:
        print(f"   ‚úÖ All predictions have confidence ‚â• {low_confidence_threshold}")
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'roc_auc': roc_auc,
        'confusion_matrix': cm,
        'prediction_confidence': prediction_confidence
    }

# Evaluate all models
iris_eval = evaluate_knn_model(iris_results, iris.target_names)
wine_eval = evaluate_knn_model(wine_results, wine.target_names)
digits_eval = evaluate_knn_model(digits_results)

# Comprehensive Visualization of Results
print(f"\nüìä COMPREHENSIVE EVALUATION VISUALIZATION")
print("=" * 60)

fig, axes = plt.subplots(3, 3, figsize=(18, 15))

# Row 1: Confusion Matrices
datasets = [iris_results, wine_results, digits_results]
evaluations = [iris_eval, wine_eval, digits_eval]
target_names_list = [iris.target_names, wine.target_names, [str(i) for i in range(10)]]

for i, (results, eval_results, target_names) in enumerate(zip(datasets, evaluations, target_names_list)):
    cm = eval_results['confusion_matrix']
    
    # Plot confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=target_names, yticklabels=target_names,
                ax=axes[0, i])
    axes[0, i].set_title(f'{results["dataset_name"]} Confusion Matrix', fontweight='bold')
    axes[0, i].set_xlabel('Predicted')
    axes[0, i].set_ylabel('Actual')

# Row 2: Prediction Confidence Distributions
for i, (results, eval_results) in enumerate(zip(datasets, evaluations)):
    confidence = eval_results['prediction_confidence']
    
    axes[1, i].hist(confidence, bins=20, alpha=0.7, edgecolor='black', color='skyblue')
    axes[1, i].axvline(confidence.mean(), color='red', linestyle='--', 
                      label=f'Mean: {confidence.mean():.3f}')
    axes[1, i].set_xlabel('Prediction Confidence')
    axes[1, i].set_ylabel('Frequency')
    axes[1, i].set_title(f'{results["dataset_name"]} Prediction Confidence', fontweight='bold')
    axes[1, i].legend()
    axes[1, i].grid(True, alpha=0.3)

# Row 3: Performance Metrics Comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
iris_scores = [iris_eval['accuracy'], iris_eval['precision'], iris_eval['recall'], 
               iris_eval['f1'], iris_eval['roc_auc']]
wine_scores = [wine_eval['accuracy'], wine_eval['precision'], wine_eval['recall'], 
               wine_eval['f1'], wine_eval['roc_auc']]
digits_scores = [digits_eval['accuracy'], digits_eval['precision'], digits_eval['recall'], 
                 digits_eval['f1'], digits_eval['roc_auc']]

x = np.arange(len(metrics))
width = 0.25

axes[2, 0].bar(x - width, iris_scores, width, label='Iris', alpha=0.8, color='blue')
axes[2, 0].bar(x, wine_scores, width, label='Wine', alpha=0.8, color='red')
axes[2, 0].bar(x + width, digits_scores, width, label='Digits', alpha=0.8, color='green')

axes[2, 0].set_xlabel('Metrics')
axes[2, 0].set_ylabel('Score')
axes[2, 0].set_title('Performance Metrics Comparison', fontweight='bold')
axes[2, 0].set_xticks(x)
axes[2, 0].set_xticklabels(metrics, rotation=45, ha='right')
axes[2, 0].legend()
axes[2, 0].grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, (iris_score, wine_score, digits_score) in enumerate(zip(iris_scores, wine_scores, digits_scores)):
    axes[2, 0].text(i - width, iris_score + 0.01, f'{iris_score:.3f}', 
                   ha='center', va='bottom', fontsize=8)
    axes[2, 0].text(i, wine_score + 0.01, f'{wine_score:.3f}', 
                   ha='center', va='bottom', fontsize=8)
    axes[2, 0].text(i + width, digits_score + 0.01, f'{digits_score:.3f}', 
                   ha='center', va='bottom', fontsize=8)

# Performance vs Complexity Analysis (use remaining subplots)
complexity_data = []
for results in datasets:
    n_train = results['X_train_scaled'].shape[0]
    n_features = results['X_train_scaled'].shape[1]
    k = results['optimal_k']
    accuracy = results['final_accuracy']
    
    complexity_data.append({
        'dataset': results['dataset_name'],
        'n_train': n_train,
        'n_features': n_features,
        'k': k,
        'accuracy': accuracy,
        'complexity_score': n_train * n_features  # Simple complexity measure
    })

complexity_df = pd.DataFrame(complexity_data)

# Scatter plot: Complexity vs Accuracy
axes[2, 1].scatter(complexity_df['complexity_score'], complexity_df['accuracy'], 
                  s=100, alpha=0.7, c=['blue', 'red', 'green'])

for i, row in complexity_df.iterrows():
    axes[2, 1].annotate(row['dataset'], 
                       (row['complexity_score'], row['accuracy']),
                       xytext=(5, 5), textcoords='offset points')

axes[2, 1].set_xlabel('Complexity Score (n_train √ó n_features)')
axes[2, 1].set_ylabel('Accuracy')
axes[2, 1].set_title('Accuracy vs Computational Complexity', fontweight='bold')
axes[2, 1].grid(True, alpha=0.3)

# Feature dimensionality impact
axes[2, 2].bar(complexity_df['dataset'], complexity_df['n_features'], 
              color=['blue', 'red', 'green'], alpha=0.7, edgecolor='black')
axes[2, 2].set_xlabel('Dataset')
axes[2, 2].set_ylabel('Number of Features')
axes[2, 2].set_title('Feature Dimensionality Comparison', fontweight='bold')
axes[2, 2].grid(True, alpha=0.3, axis='y')

# Add value labels
for i, (dataset, n_features) in enumerate(zip(complexity_df['dataset'], complexity_df['n_features'])):
    axes[2, 2].text(i, n_features + 1, str(n_features), 
                   ha='center', va='bottom', fontweight='bold')

plt.suptitle('Comprehensive KNN Model Evaluation', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Final Summary Report
print(f"\nüìã FINAL EVALUATION SUMMARY:")
print("=" * 60)

summary_data = []
for results, eval_results in zip(datasets, evaluations):
    summary_data.append({
        'Dataset': results['dataset_name'],
        'Samples': results['X_train_scaled'].shape[0] + results['X_test_scaled'].shape[0],
        'Features': results['X_train_scaled'].shape[1],
        'Classes': len(np.unique(results['y_test'])),
        'Optimal K': results['optimal_k'],
        'Best Metric': results['best_metric'],
        'Accuracy': f"{eval_results['accuracy']:.4f}",
        'F1-Score': f"{eval_results['f1']:.4f}",
        'ROC-AUC': f"{eval_results['roc_auc']:.4f}"
    })

final_summary = pd.DataFrame(summary_data)
print(final_summary.to_string(index=False))

print(f"\nüèÜ PERFORMANCE RANKING:")
print("-" * 30)
performance_ranking = sorted(evaluations, key=lambda x: x['accuracy'], reverse=True)
dataset_ranking = [datasets[evaluations.index(eval_r)]['dataset_name'] for eval_r in performance_ranking]

for i, (dataset, eval_r) in enumerate(zip(dataset_ranking, performance_ranking), 1):
    print(f"{i}. {dataset}: {eval_r['accuracy']:.4f} accuracy")

print(f"\nüîç KEY INSIGHTS:")
print("-" * 20)
print(f"1. {dataset_ranking[0]} dataset shows best KNN performance")
print(f"2. Feature scaling was critical for all datasets")
print(f"3. Parameter optimization improved performance significantly")
print(f"4. Higher dimensionality (Digits) shows impact of curse of dimensionality")
print(f"5. KNN demonstrates strong performance across diverse problem types")
````

````python
# Decision Boundary Visualization (Iris Dataset)
print("üé® DECISION BOUNDARY VISUALIZATION")
print("=" * 50)

# For visualization, we'll use only 2 features from Iris dataset
# This allows us to create 2D decision boundary plots

print("üìä Creating 2D decision boundary visualization using Iris dataset...")
print("Using first two features: Sepal Length vs Sepal Width")

# Prepare 2D data for visualization
X_iris_2d = iris.data[:, :2]  # Use only first 2 features
y_iris = iris.target

# Split and scale the 2D data
X_train_2d, X_test_2d, y_train_2d, y_test_2d = train_test_split(
    X_iris_2d, y_iris, test_size=0.2, random_state=42, stratify=y_iris
)

scaler_2d = StandardScaler()
X_train_2d_scaled = scaler_2d.fit_transform(X_train_2d)
X_test_2d_scaled = scaler_2d.transform(X_test_2d)

# Train KNN models with different K values for boundary visualization
k_values_viz = [1, 3, 5, 15]
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.ravel()

def plot_decision_boundary(X, y, model, scaler, ax, title, feature_names):
    """Plot decision boundary for 2D classification"""
    
    # Create a mesh grid
    h = 0.02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                        np.arange(y_min, y_max, h))
    
    # Make predictions on the mesh grid
    mesh_points = np.c_[xx.ravel(), yy.ravel()]
    mesh_points_scaled = scaler.transform(mesh_points)
    Z = model.predict(mesh_points_scaled)
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary
    ax.contourf(xx, yy, Z, alpha=0.4, cmap='viridis')
    ax.contour(xx, yy, Z, colors='black', linewidths=0.5, alpha=0.5)
    
    # Plot data points
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', 
                        edgecolors='black', s=50, alpha=0.8)
    
    ax.set_xlabel(feature_names[0])
    ax.set_ylabel(feature_names[1])
    ax.set_title(title, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    return scatter

# Create decision boundary plots for different K values
feature_names_2d = [iris.feature_names[0], iris.feature_names[1]]

for i, k in enumerate(k_values_viz):
    # Train KNN model with specific K
    knn_viz = KNeighborsClassifier(n_neighbors=k)
    knn_viz.fit(X_train_2d_scaled, y_train_2d)
    
    # Calculate accuracy for this K
    y_pred_2d = knn_viz.predict(X_test_2d_scaled)
    accuracy_2d = accuracy_score(y_test_2d, y_pred_2d)
    
    # Plot decision boundary
    scatter = plot_decision_boundary(
        X_train_2d, y_train_2d, knn_viz, scaler_2d, axes[i],
        f'KNN Decision Boundary (K={k})\nAccuracy: {accuracy_2d:.3f}',
        feature_names_2d
    )

# Add colorbar
plt.colorbar(scatter, ax=axes, orientation='horizontal', 
             fraction=0.05, pad=0.1, ticks=[0, 1, 2],
             label='Iris Species (0: setosa, 1: versicolor, 2: virginica)')

plt.suptitle('KNN Decision Boundaries: Effect of K Parameter\n(Iris Dataset - First 2 Features)', 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Analysis of decision boundary complexity
print(f"\nüîç DECISION BOUNDARY ANALYSIS:")
print("-" * 50)

boundary_analysis = []
for k in k_values_viz:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_2d_scaled, y_train_2d)
    
    # Test accuracy
    test_acc = accuracy_score(y_test_2d, knn_temp.predict(X_test_2d_scaled))
    
    # Training accuracy
    train_acc = accuracy_score(y_train_2d, knn_temp.predict(X_train_2d_scaled))
    
    boundary_analysis.append({
        'K': k,
        'Train Accuracy': f"{train_acc:.4f}",
        'Test Accuracy': f"{test_acc:.4f}",
        'Overfitting': f"{train_acc - test_acc:+.4f}",
        'Boundary Complexity': 'High' if k <= 3 else 'Medium' if k <= 10 else 'Low'
    })

boundary_df = pd.DataFrame(boundary_analysis)
print(boundary_df.to_string(index=False))

print(f"\nüìã BOUNDARY COMPLEXITY INSIGHTS:")
print(f"   ‚Ä¢ K=1: Highly complex, jagged boundaries (potential overfitting)")
print(f"   ‚Ä¢ K=3: Moderately complex, captures local patterns")
print(f"   ‚Ä¢ K=5: Balanced complexity, smooth but adaptive boundaries")
print(f"   ‚Ä¢ K=15: Simple, smooth boundaries (potential underfitting)")

print(f"\nüéØ OPTIMAL K SELECTION CONSIDERATIONS:")
print(f"   ‚Ä¢ Lower K: More flexible, higher variance, potential overfitting")
print(f"   ‚Ä¢ Higher K: More stable, higher bias, potential underfitting")
print(f"   ‚Ä¢ Cross-validation helps find the sweet spot")
print(f"   ‚Ä¢ Visual inspection of boundaries aids understanding")

# Neighborhood illustration for a specific test point
print(f"\nüîç NEIGHBORHOOD ILLUSTRATION:")
print("-" * 40)

# Select a test point for neighborhood analysis
test_point_idx = 0
test_point = X_test_2d_scaled[test_point_idx:test_point_idx+1]
test_point_original = X_test_2d[test_point_idx:test_point_idx+1]
true_label = y_test_2d[test_point_idx]

print(f"Analyzing neighborhood for test point:")
print(f"   ‚Ä¢ Original coordinates: ({test_point_original[0,0]:.2f}, {test_point_original[0,1]:.2f})")
print(f"   ‚Ä¢ True class: {iris.target_names[true_label]}")

# Find K nearest neighbors for different K values
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for plot_idx, k in enumerate([3, 7]):
    knn_neighbors = KNeighborsClassifier(n_neighbors=k)
    knn_neighbors.fit(X_train_2d_scaled, y_train_2d)
    
    # Find distances and indices of neighbors
    distances, indices = knn_neighbors.kneighbors(test_point)
    neighbor_points = X_train_2d_scaled[indices[0]]
    neighbor_labels = y_train_2d[indices[0]]
    
    # Make prediction
    prediction = knn_neighbors.predict(test_point)[0]
    prediction_proba = knn_neighbors.predict_proba(test_point)[0]
    
    # Plot training data
    scatter = axes[plot_idx].scatter(X_train_2d_scaled[:, 0], X_train_2d_scaled[:, 1], 
                                   c=y_train_2d, cmap='viridis', alpha=0.6, s=30)
    
    # Highlight neighbors
    axes[plot_idx].scatter(neighbor_points[:, 0], neighbor_points[:, 1], 
                         c=neighbor_labels, cmap='viridis', s=150, 
                         edgecolors='red', linewidths=3, alpha=0.9,
                         label=f'K={k} Neighbors')
    
    # Highlight test point
    axes[plot_idx].scatter(test_point[0, 0], test_point[0, 1], 
                         c='black', s=200, marker='*', 
                         edgecolors='white', linewidths=2,
                         label='Test Point')
    
    # Draw circles to show neighborhood
    for i, (distance, neighbor_point) in enumerate(zip(distances[0], neighbor_points)):
        if i == k-1:  # Draw circle for the farthest neighbor
            circle = plt.Circle((test_point[0, 0], test_point[0, 1]), distance, 
                              fill=False, color='red', linestyle='--', linewidth=2)
            axes[plot_idx].add_patch(circle)
    
    axes[plot_idx].set_xlabel(feature_names_2d[0] + ' (scaled)')
    axes[plot_idx].set_ylabel(feature_names_2d[1] + ' (scaled)')
    axes[plot_idx].set_title(f'K={k} Nearest Neighbors\nPrediction: {iris.target_names[prediction]}\nConfidence: {prediction_proba.max():.3f}', 
                           fontweight='bold')
    axes[plot_idx].legend()
    axes[plot_idx].grid(True, alpha=0.3)

plt.suptitle('KNN Neighborhood Analysis for Test Point', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\nüí° NEIGHBORHOOD INSIGHTS:")
print(f"   ‚Ä¢ Larger K includes more distant neighbors")
print(f"   ‚Ä¢ Decision is based on majority vote within the neighborhood")
print(f"   ‚Ä¢ Circle radius shows the distance to the K-th nearest neighbor")
print(f"   ‚Ä¢ Different K values can lead to different predictions")
print(f"   ‚Ä¢ Visualization helps understand the local decision process")
````
